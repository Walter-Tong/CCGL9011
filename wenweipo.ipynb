```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import time
import requests
import random
import os

#for Wen Wei Po they don't storge the date in the url, so just change the day mamual in the pop up window.
```

```python
selenium_service = Service('path/to/chromedriver')
chrome_options = Options()
chrome_options.add_argument("--headless")  # Run in headless mode
driver = webdriver.Chrome()
```

```python
url = "https://www.wenweipo.com/todaywenwei"
driver.get(url)
```

```python
date = "2020-11-02"
```

```python
click_element = driver.find_element(By.CSS_SELECTOR, "div.searchMore.da-bre")

driver.implicitly_wait(1)

time.sleep(5)

click_element.click()

time.sleep(10)

while (driver.find_element(By.CSS_SELECTOR, "div.searchMore.da-bre").find_element(By.CSS_SELECTOR, "a").text == "加載更多"):
    click_element = driver.find_element(By.CSS_SELECTOR, "div.searchMore.da-bre")

    driver.implicitly_wait(1)

    time.sleep(5)

    click_element.click()

    time.sleep(10)
```

```python
links = []
```

```python
soup = BeautifulSoup(driver.page_source, "html.parser")

soup
```

```python
content = soup.find_all(class_="searchItem")

for i in content:
    links.append(i.find("a").get("href"))

```

```python
os.makedirs("wenweipo/" + date+ "/article")
```

```python
for i in links:
    driver.get(i)

    article = BeautifulSoup(driver.page_source, "html.parser")

    title = article.find(class_="post-header").find("h1").text

    aspect = article.find(class_="post-header").find("img").get("alt")

    date_and_time = article.find(class_="post-header").find("time").text

    content = [j.text for j in article.find_all("p")]

    file_name = title.replace(" ", "") + ".txt"

    file = open("wenweipo/"+ date  + "/article/" + file_name, "w", encoding='utf-8')

    file.write(title + "\n")

    file.write(aspect + "\n")

    file.write(date_and_time + "\n")

    for j in content:
        file.write(j + "\n")

    file.close()

    time.sleep(30*random.uniform(1, 2))

```
