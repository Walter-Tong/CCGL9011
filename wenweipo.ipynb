```python
links = []
topics = []
```

```python
soup = BeautifulSoup(driver.page_source, "html.parser")

soup
```

```python
content = soup.find_all(class_="searchItem")

for i in content:
    links.append(i.find("a").get("href"))
    topics.append(i.find("a").text)
```

```python
if len(links) > 40:
    links = random.sample(links, 40)
```

```python
links = []

file_name = date + "links.txt"

file = open(direct + date + "/" + file_name, "r", encoding='utf-8')

for i in file:
    links.append(i.strip())

links
```

```python
proxy_address = 'brd.superproxy.io'
proxy_port = 22225
proxy_username = 'brd-customer-hl_4abea413-zone-data_center'
proxy_password = 'lmz4ftfwixm4'

proxies = {
    'http': f'http://{proxy_username}:{proxy_password}@{proxy_address}:{proxy_port}',
    'https': f'https://{proxy_username}:{proxy_password}@{proxy_address}:{proxy_port}'
}

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.93 Safari/537.36',
    'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8',
    'Accept-Language': 'en-US,en;q=0.9',
    'Accept-Encoding': 'gzip, deflate, br',
    'Referer': 'https://www.google.com/',
    'Connection': 'keep-alive',
    # Add any other headers you want to include
}
```

```python
count = 0

for i in links:
    if os.path.exists(direct + date + "/article/" + str(count) + ".txt"):
        count += 1
        continue

    reponse = requests.get(i, proxies=proxies, headers=headers)

    article = BeautifulSoup(reponse.content, "html.parser")

    title = article.find(class_="post-header").find("h1").text

    aspect = article.find(class_="post-header").find("img").get("alt")

    date_and_time = article.find(class_="post-header").find("time").text

    content = [j.text for j in article.find_all("p")]

    file_name = str(count) + ".txt"

    file = open("wenweipo/"+ date  + "/article/" + file_name, "w", encoding='utf-8')

    file.write(title + "\n")

    file.write(aspect + "\n")

    file.write(date_and_time + "\n")

    for j in content:
        file.write(j + "\n")

    file.close()

    time.sleep(30*random.uniform(1, 2))

    count += 1

```
