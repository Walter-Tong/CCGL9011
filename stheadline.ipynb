```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import time
import requests
import random
import os
```

```python
url = "https://std.stheadline.com/daily/formerly/%E6%97%A5%E5%A0%B1-%E6%98%94%E6%97%A5-2020-11-02"

response = requests.get(url)
```

```python
soup = BeautifulSoup(response.content, "html.parser")

soup
```

```python
links = []
```

```python
content = soup.find_all(class_="media-body")

content[1]
```

```python
for i in content[1:]:
    links.append(i.find("a").get("href"))

links
```

```python
os.makedirs("stheadline/" + url[-10:]+ "/article")
```

```python
selenium_service = Service('path/to/chromedriver')
chrome_options = Options()
chrome_options.add_argument("--headless")  # Run in headless mode
driver = webdriver.Chrome()
```

```python
for i in links:
    driver.get(i)

    article = BeautifulSoup(driver.page_source, "html.parser")

    title = article.find("article").find("h1").text

    aspect = i.split("-")[1]

    content = []

    for j in article.find(class_="video paragraphs medium").find_all("p"):
        content.append(j.text)

    file_name = title.replace(" ", "") + ".txt"

    file = open("stheadline/"+ url[-10:] + "/article/" + file_name, "w", encoding='utf-8')

    file.write(title + "\n")

    file.write(aspect + "\n")

    file.write(url[-10:] + "\n")

    for j in content:
        file.write(j + "\n")

    file.close()

    time.sleep(30*random.uniform(1, 2))

```
