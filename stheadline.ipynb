```python
#read from links.txt

links = []

file_name = url[-10:] + "links.txt"

file = open("stheadline/"+ url[-10:] + "/" + file_name, "r", encoding='utf-8')

for i in file:
    links.append(i.strip())

links
```

```python
count = 0

for i in links:
    if os.path.exists("stheadline/"+ url[-10:] + "/article/" + str(count) + ".txt"):
        count += 1
        continue

    response = requests.get(i, proxies=proxies, headers=headers)

    article = BeautifulSoup(response.content, "html.parser")

    title = article.find("article").find("h1").text

    aspect = i.split("-")[1]

    content = []

    for j in [k.text for k in article.find_all("p")[1:-1]]:
        content.append(j)

    file_name = str(count) + ".txt"

    file = open("stheadline/"+ url[-10:] + "/article/" + file_name, "w", encoding='utf-8')

    file.write(title + "\n")

    file.write(aspect + "\n")

    file.write(url[-10:] + "\n")

    for j in content:
        file.write(j + "\n")

    file.write(i)

    file.close()

    time.sleep(30*random.uniform(1, 1.5))

    count += 1

```
