```python
import calendar
import random
from datetime import datetime


start_year = 2020
start_month = 11

end_year = 2023
end_month = 10

cur_year = start_year
cur_month = start_month

dates = []

count = 0

count2 = 0

while (cur_year < end_year or cur_month < end_month):
    date_range = calendar.monthrange(cur_year, cur_month)[1]
    while (count2 < count * 4 + 4):
        date_draw = random.randint(1, date_range)

        if (calendar.weekday(cur_year, cur_month, date_draw) >= 5):
            continue

        to_add = formatted_date = datetime.strftime(datetime.strptime(str(cur_year) + "-" + str(cur_month) + "-" + str(date_draw), "%Y-%m-%d"), "%Y-%m-%d")

        if (not(len([ i for i in dates if i == to_add]))):
            dates.append(to_add)
            count2 += 1
        
    cur_month = cur_month%12 + 1

    count += 1

    if (cur_month == 1):
        cur_year += 1

dates.sort()

dates
```
