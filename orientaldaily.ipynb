```python
soup = BeautifulSoup(response.content, "html.parser")

soup
```

```python
base_url = "https://orientaldaily.on.cc"

links = []
topics = []
```

```python
content = soup.find_all(class_="clearList")

for i in content:
    for j in i.findAll("a"):
        links.append(base_url + j.get("href"))
        topics.append(j.text)

links
topics
```

```python
folder_path = "orientaldaily/" + date + "/article"

if not os.path.exists(folder_path):
    os.makedirs(folder_path)
```

```python
file_name = date + "links.txt"

file = open("orientaldaily/"+ date + "/" + file_name, "w", encoding='utf-8')

for i in links:
    file.write(i + "\n")

file.close()
```

```python
file_name = date + "topics.txt"

file = open("orientaldaily/"+ date + "/" + file_name, "w", encoding='utf-8')

for i in topics:
    file.write(i + "\n")

file.close()
```

```python
links = []

file_name = date + "links.txt"

file = open(direct + date + "/" + file_name, "r", encoding='utf-8')

for i in file:
    links.append(i.strip())

links
```

```python
count = 0

for i in links:
    if os.path.exists(direct + date + "/article/" + str(count) + ".txt"):
        count += 1
        continue

    response = requests.get(i, proxies=proxies)

    article = BeautifulSoup(response.content, "html.parser")

    title = article.find("h1").text

    aspect = ""

    if (not(article.find('link', rel="alternate", type="application/rss+xml"))):
        count += 1
        continue
    else:
        aspect = article.find('link', rel="alternate", type="application/rss+xml").get("title").split()[0]

    content = []

    content.append(article.find(class_="leadin").find("p").text)

    for j in article.find_all(class_="caption"):
        content.append(j.text)

    for j in article.find(id="contentCTN-right").find_all("p"):
        content.append(j.text)

    for j in [k.find("li") for k in article.find(id="contentCTN-right").find_all("ul")]:
        content.append(j.text)

    for j in article.find(id="contentCTN-right").find_all("h3"):
        content.append(j.text)

    file_name =  str(count) + ".txt"

    file = open("orientaldaily/" + date + "/article/" + file_name, "w", encoding='utf-8')

    file.write(title + "\n")

    file.write(aspect + "\n")

    file.write(date + "\n")

    for j in content:
        file.write(j + "\n")

    file.write(i)

    file.close()

    time.sleep(30*random.uniform(1, 1.5))

    count += 1
```
